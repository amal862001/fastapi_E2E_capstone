```python
import numpy as np
from sqlalchemy import engine
```

```python
import pandas as pd
df = pd.read_csv("nyc_311_requests.csv")
print(f"Total rows: {df.shape[0]}")
df.head()
```

### Data preprocessing 

```python
# 

df = df[[
    'unique_key',
    'created_date',
    'closed_date',
    'agency',
    'agency_name',
    'complaint_type',
    'descriptor',
    'location_type',
    'incident_zip',
    'city',
    'borough',
    'status',
    'resolution_description',
    'latitude',
    'longitude',
    'resolution_action_updated_date'
]]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['city'] = df['city'].str.upper().str.strip()
```

```python
df['city'].unique()
```

```python
df['incident_zip'].unique()
```

```python
print((df['incident_zip'] == 'N/A').sum())
```

```python
print(df['borough'].value_counts(dropna=False))
```

```python
# Replace 'Unspecified' and empty strings with None
df['borough'] = df['borough'].replace({"Unspecified":np.nan})
df['borough'] = df['borough'].str.upper().str.strip()
```

```python
df['borough'].value_counts(dropna=False)
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
# A complaint with no borough is essentially an incomplete record

df = df.dropna(subset=['borough'])
```

```python
non_nullable_columns = [
    'unique_key',
    'created_date',
    'agency',
    'agency_name',
    'complaint_type',
    'borough',
    'status'
]

print(df[non_nullable_columns].isnull().sum())
```

```python
# Fix dates 
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')
df['closed_date'] = pd.to_datetime(df['closed_date'], errors='coerce')
df['resolution_action_updated_date'] = pd.to_datetime(df['resolution_action_updated_date'], errors='coerce')

# Fix ZIP (convert float back to clean string) 
df['incident_zip'] = df['incident_zip'].apply(
    lambda x: str(int(x)).zfill(5) if pd.notnull(x) else None
)

# Verify 
print(df.dtypes)
```

```python
invalid_zips = ['00000', 'N/A', '', ' ']
for val in invalid_zips:
    count = (df['incident_zip'] == val).sum()
    print(f"'{val}': {count} rows")
```
